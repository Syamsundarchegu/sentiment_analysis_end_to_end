```python
import pandas as pd
from dotenv import load_env, find_env
import os
_ = load_env(find_env())



df = pd.read_csv('twitter_training.csv', encoding='utf-8')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
import pymongo


try:
    client = pymongo.MongoClient(host=os.environ['mongo_url'])
    print("mongodb connection sucessfully established")
except Exception as e:
    print("Failed to connect to MongoDB",e)
```

```python
db = client['twitter']
collection = db['tweets']
```

```python
collection.insert_many(df.to_dict(orient="records"))
```

```python
df = pd.DataFrame(list(collection.find()))
```

```python
df.columns
```

```python
df.drop(columns=['_id'],axis=1,inplace=True)
```
